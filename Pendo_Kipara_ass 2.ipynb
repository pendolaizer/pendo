Question no 2 data structure

```python
def list_operation(numbers):
    if not numbers:
        return (0, None, None)

    total_sum = sum(numbers)
    max_number = max(numbers)
    min_number = min(numbers)

    return (total_sum, max_number, min_number)


numbers = [1, 2, 3, 4, 5]
result = list_operation(numbers)
print("Sum:", result[0], "Max:", result[1], "Min:", result[2])
```

```python
#Function dict_operations(students_dict)
def dict_operations(students):
    return [name for name, grade in students.items() if grade >= 80]

students = {
    "Jessy": 85,
    "Alex": 90,
    "Betty": 75,
    "Edna": 95
}
print(dict_operations(students))
```

```python
#Function set_operations(list1, list2)
def set_operations(list1, list2):
    set1 = set(list1)
    set2 = set(list2)
    return set1.intersection(set2)
print(set_operations([1, 2, 3], [2, 3, 4]))
print(set_operations([1, 2], [3, 4, 5]))
```

```python
#Function arithmetic_ops(a, b)
def arithmetic_ops(a, b):
    return {
        "sum": a + b,
        "difference": a - b,
        "product": a * b,
        "quotient": a / b
    }
print(arithmetic_ops(200, 5))
```

```python
#Function bitwise_ops(a, b)
def bitwise_ops(a, b):
    return {
        "and": a & b,
        "or": a | b,
        "xor": a ^ b
    }
print(bitwise_ops(22, 10))
```
